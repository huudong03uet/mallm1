```python
from mallm import scheduler
from mallm.utils.config import Config
```

```python

mallm_scheduler = scheduler.Scheduler(
    Config(
        input_json_file_path="data/datasets/etpc_debugging.json",
        output_json_file_path="test_out.json",
        task_instruction_prompt="Paraphrase the input text.",
        endpoint_url="http://192.168.9.88:9876/v1",
        model_name='GPT-OSS-120B'
    )
)
mallm_scheduler.run()
```

```python
# read file test_out.json
import json
with open('test_out.json', 'r') as f:
    datas = json.load(f)
```

```python
# read file /home/sysadmin/workspace/dongnh/test_all/test2/mallm/data/COLIEE2026/dataset/All/All/task4/answer-task4
# R07-01-O	Y
# R07-02-O	N
# R07-02-U	Y
with open('/home/sysadmin/workspace/dongnh/test_all/test2/mallm/data/COLIEE2026/dataset/All/All/task4/answer-task4', 'r') as f:
    
    data_label = f.read().splitlines()
data_label_dict = {}
for line in data_label:
    example_id, label = line.split('\t')
    data_label_dict[example_id] = label
    
data_label_dict
```

```python
count_correct = 0
for data in datas:
    # print(f"Final Answer: {data['finalAnswer'].split()[-1]}")
    final_answer = data['finalAnswer'].split()[-1]
    assert final_answer == "Yes" or final_answer == "No", f"Unexpected final answer: {final_answer}"
    # print(f"Reference: {data['references'][0]}")
    id = data['exampleId']
    referece = "Yes" if data_label_dict[id] == "Y" else "No"
    # count_correct += 1 if final_answer == referece else 0
    if final_answer == referece:
        count_correct += 1
    else:
        print(f"Incorrect example {data['exampleId']}: Final Answer: {final_answer}, Reference: {referece}")

print(f"Accuracy: {count_correct / len(datas) * 100:.2f}%")
```
